```python
import numpy as np
import matplotlib.pyplot as plt

y = np.array([1.6, 0.8, 1.2, 0.5, 0.9, 1.1, 1.1, 0.6, 1.5, 0.8, 0.9, 1.2, 0.5, 1.3, 0.8, 1.2])
n = len(y)

plt.figure(figsize=(10,4))
plt.plot(range(1, n+1), y, marker='o')
plt.title(" ")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.grid(True)
plt.show()

y_t = y[:-1]
y_t1 = y[1:]

plt.figure(figsize=(6,6))
plt.scatter(y_t, y_t1)
plt.title(" y(t+1)  y(t)")
plt.xlabel("y(t)")
plt.ylabel("y(t+1)")
plt.grid(True)
plt.show()

y_mean = np.mean(y)
numerator = np.sum((y_t - y_mean)*(y_t1 - y_mean))
denominator = np.sum((y - y_mean)**2)
r1 = numerator / denominator
print(f"    r1 = {r1:.3f}")
#  2
np.random.seed(42)
n2 = 20
trend = 0.1
epsilon = np.random.normal(0, 1, n2)
y_rw = np.zeros(n2)
y_rw[0] = 0
for t in range(1, n2):
    y_rw[t] = y_rw[t-1] + trend + epsilon[t]

tau = 1
y_forecast = y_rw[-1] + trend*tau
print(f" y(t+{tau}) = {y_forecast:.3f}")

y_actual = y_rw[-1] + trend + epsilon[-1]
forecast_error = y_actual - y_forecast
print(f"  e(t+{tau}) = {forecast_error:.3f}")

mse = np.var(epsilon)
print(f"C   MSE = {mse:.3f}")

plt.figure(figsize=(10,4))
plt.plot(range(1,n2+1), y_rw, marker='o', label="   ")
plt.title("   ")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.grid(True)
plt.legend()
plt.show()
```
